## B.2 Acerca de los comandos mágicos
Los comandos especiales de IPython (que no están integrados en el propio Python) se conocen como `comandos mágicos`. Están diseñados para facilitar tareas habituales y permitirte controlar fácilmente el comportamiento del sistema IPython. Un comando mágico es cualquier comando precedido por el símbolo de porcentaje %. Por ejemplo, puedes comprobar el tiempo de ejecución de cualquier instrucción de Python, como una multiplicación de matrices, utilizando la función mágica %timeit:

### Table B-2. Some frequently used IPython magic commands

| Command | Description |
| :--- | :--- |
| `%quickref` | Muestra la tarjeta de referencia rápida de IPython (Quick Reference Card). |
| `%magic` | Muestra la documentación detallada de todos los comandos mágicos disponibles. |
| `%debug` | Entra al depurador interactivo en la parte inferior del historial de ejecución (traceback) de la última excepción. |
| `%hist` | Imprime el historial de comandos de entrada (y opcionalmente de salida). |
| `%pdb` | Entra automáticamente al depurador después de cualquier excepción. |
| `%paste` | Ejecuta código de Python preformateado desde el portapapeles. |
| `%cpaste` | Abre un indicador (prompt) especial para pegar manualmente código de Python que se va a ejecutar. |
| `%reset` | Elimina todas las variables/nombres definidos en un espacio de nombres (namespace) interactivo. |
| `%page OBJECT` | Muestra el objeto de forma estética (pretty-print) a través de un paginador. |
| `%run script.py` | Ejecuta un script de Python dentro de IPython. |
| `%prun statement` | Ejecuta una declaración (`statement`) con `cProfile` y reporta la salida del perfilador. |
| `%time statement` | Reporta el tiempo de ejecución de una sola declaración. |
| `%timeit statement` | Ejecuta una declaración múltiples veces para calcular un promedio general del tiempo de ejecución; es útil para medir el tiempo de código con un tiempo de ejecución muy corto. |
| `%who`, `%who_ls`, `%whos` | Muestra las variables definidas en el espacio de nombres interactivo, con niveles variables de información/detalle. |
| `%xdel variable` | Elimina una variable e intenta borrar cualquier referencia al objeto en los elementos internos de IPython. |
